```julia
function model3opt(max_units = 25, min_dist = 1, minimum_density = 0.025)
    
    model3 = Model(Cbc.Optimizer)

    
    #Here we call the data and electric grid CSVs
    densityCSV = CSV.read("data.csv", copycols = true)
    gridCSV = CSV.read("electricity.csv", copycols = true)
    data_to_dist = CSV.read("Data_With_Dist.csv", copycols = true)
    
    #We want to have a value greater than the minimum acceptable density and filter density CSV
    filter_density = filter(row -> row[:value] > minimum_density, densityCSV)
    
    #We want only the first column and we want the size of that... shoutout s083
    density = size(filter_density)[1]
    elec_grid = size(gridCSV)[1]
    data_dist = size(data_to_dist)[1]
    
    @variable(model3, b[1:density], bin)
    
    for i = 1:data_dist
        @constraint(model3, data_to_dist[9][i] >= min_dist*b[i])
    end
    
    @objective(model3, Max, sum((filter_density[9][i] + (((1/2)^(data_dist[9][i]))*filter_density[9][i]))*b[i] for i=1:density))
    
    optimize!(model3)
    objective = objective_value(model3)
    
    soln = value.(b)
    sol = [i for (i,n) in enumerate(soln) if n == 1]
    
    return (max_units, min_dist, filter_density, obj_val, length(sol), sol)
    
end    

    
```

```julia
using Dates
using DataFrames
```

```julia
pop_density_list = [0.005 0.01 0.025]

num_hubs = [5 10 20 50]

mini_distance = [0.1 0.5 1 1.5]

results = []

for each_value in pop_density_list
    println("Population Density: ", each_value, Dates.Time(Dates.now()))
    
    for each_hub in num_hubs
        println("Number of Hubs: ", each_hub, Dates.Time(Dates.now()))
        
        for each_distance in mini_distance
            println("Minimum Distances: ", each_distance, Dates.Time(Dates.now()))
            push!(results, model3opt(each_value, each_hub, each_distance))
        end
    end
end

resultsCSV = DataFrame(results)

CSV.write("Model_3.csv", resultsCSV)
```
